```python
import datetime
import numpy as np
import pandas as pd 
import seaborn
import matplotlib.pyplot as plt 
import matplotlib.dates as dates
%matplotlib inline
```

```python
base = datetime.datetime(2014, 1, 1)
year = np.array([base + datetime.timedelta(days=i) for i in xrange(365)])
```

```python
df = pd.read_csv('BOSTON.csv', usecols=['OPEN_DT', 'CASE_TITLE'], parse_dates='OPEN_DT')
```

```python
df_sig = df.loc[df['CASE_TITLE'] == 'Signal/Sign Repair']
df_build = df.loc[df['CASE_TITLE'] == 'Building Conditions']
df_living = df.loc[df['CASE_TITLE'] == 'Living Conditions']
df_street = df.loc[df['CASE_TITLE'] == 'Street/Sidewalk Maintenance']
df_general = df.loc[df['CASE_TITLE'] == 'General Request or Complaint']
df_park = df.loc[df['CASE_TITLE'] == 'Parking Enforcement']
```

```python
df = df
```

```python
date_list = []
for day in year:
    date_list.append(df_sig.loc[df_sig['OPEN_DT'] == day].groupby('OPEN_DT').agg('count'))
```
